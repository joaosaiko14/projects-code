```python
#Bibliotecas necessarias
#pip install gtts pygame pydub SpeechRecognition

from gtts import gTTS
#from pygame import mixer
import os
import time
from pydub import AudioSegment
import speech_recognition as sr
from IPython.display import Audio, display

recognizer = sr.Recognizer()

tts_pt = gTTS("olá, testando algoritmo do TDE.", lang='pt')
filename = 'audio.mp3'
tts_pt.save(filename)

#Carregando o arquivo de audio em MP3
audioMP3 = AudioSegment.from_mp3(filename)

#Convertando o arquivo MP3 carregado para WAV
audioMP3.export('audio.wav', format='wav')

# Tocando o áudio
audio_data = open(filename, "rb").read()
display(Audio(audio_data))

#após a condição apaga o arquivo de audio para melhor organização
# old_file = filename
# os.remove(old_file)

#Estrutura para escrever o audio em texto
with sr.AudioFile("audio.wav") as source:
    audio = recognizer.record(source)

    texto = recognizer.recognize_google(audio, language="pt-BR")

print(texto)

# old_file_wav = "audio.wav"
# os.remove(old_file_wav)
```
